```python
import re

class Registration:
    def __init__(self):
        self.users = {}

    def register(self, username, email, name, password, phone):
        if username in self.users:
            print("Username already exists. Please choose a different username.Start the Registration from the Begining")
        elif email in [user_info['email'] for user_info in self.users.values()]:
            print("Email address is already in use. Please use a different email.Start the Registration from the Begining.")
        elif phone in [user_info['phone'] for user_info in self.users.values()]:
            print("Phone number is already in use. Please use a different phone number.Start the Registration from the Begining.")
        elif not self.is_valid_phone(phone):
            print("Invalid phone number. Phone number must be 11 digits long, starting with '01' followed by a digit from 3 to 9.")
        elif not self.is_valid_email(email):
            print("Invalid email.")
        elif not self.is_valid_password(password):
            print("Invalid password. Password must contain at least 6 characters, including 2 digits and 2 special characters.")
        elif not re.match("^[A-Za-z ]+$", name):
            print("Invalid name. Name must contain only letters and spaces.")
        else:
            user_info = {'name': name, 'email': email, 'password': password, 'phone': phone}
            self.users[username] = user_info
            print("Registration successful. You can now log in.")

    def is_valid_password(self, password):
        if len(password) < 6:
            return False

        digit_count = len(re.findall(r'\d', password))
        special_char_count = len(re.findall(r'[!@#$%^&*]', password))

        return digit_count >= 2 and special_char_count >= 2

    def is_valid_phone(self, phone):
        return re.match(r'^01[3-9][0-9]{8}$', phone) is not None

    def is_valid_email(self, email):
        return re.match(r'\b[A-Za-z0-9._%+-]+@[A-Za-z0-9.-]+\.[A-Z|a-z]{2,7}\b', email) is not None
```
